```python
import time
from pprint import pprint
from tqdm import tqdm

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
from matplotlib.mlab import psd
from obspy import Stream, UTCDateTime, read
from obspy.clients.fdsn import Client
from obspy.signal.trigger import recursive_sta_lta, trigger_onset
```

```python
def download_geofon_data(
    network,
    station,
    location,
    channel,
    start_time,
    end_time,
    output_filename,
    chunk_size,
):
    """
    Mengunduh data seismik dari jaringan GEOFON dalam rentang waktu tertentu.

    Parameters:
        network (str): Kode jaringan (contoh: "GE").
        station (str): Kode stasiun (contoh: "BBJI").
        location (str): Lokasi (contoh: "--").
        channel (str): Kanal (contoh: "BHZ").
        start_time (str): Waktu mulai dalam format "YYYY-MM-DD HH:MM:SS".
        end_time (str): Waktu akhir dalam format "YYYY-MM-DD HH:MM:SS".
        output_filename (str): Nama file output untuk menyimpan data (format MSEED).
        chunk_size (int): Ukuran chunk dalam detik (contoh: 86400 untuk 1 hari).

    Returns:
        None
    """
    # Inisialisasi client GEOFON
    client = Client("GFZ")

    # Konversi waktu ke objek UTCDateTime
    start_time = UTCDateTime(start_time)
    end_time = UTCDateTime(end_time)

    # Buat Stream kosong untuk menampung semua data
    full_stream = Stream()

    # Download data per chunk untuk menghindari timeout
    current_time = start_time

    try:
        while current_time < end_time:
            chunk_end = min(current_time + chunk_size, end_time)

            print(f"Mengunduh data untuk: {current_time.date}")

            try:
                st = client.get_waveforms(
                    network=network,
                    station=station,
                    location=location,
                    channel=channel,
                    starttime=current_time,
                    endtime=chunk_end,
                )

                full_stream += st
                print(f"Berhasil mengunduh data untuk {current_time.date}")

            except Exception as e:
                print(f"Gagal mengunduh data untuk {current_time.date}: {str(e)}")

            # Pindah ke waktu berikutnya
            current_time += chunk_size

            # Tambahkan jeda kecil antara request
            time.sleep(1)

        print("\nProses unduh selesai!")
        print("Informasi data lengkap:")
        print(full_stream)

        # Simpan data dalam format mseed
        full_stream.write(output_filename, format="MSEED")
        print(f"Data berhasil disimpan ke {output_filename}")

    except Exception as e:
        print(f"Error utama: {str(e)}")


if __name__ == "__main__":
    # Contoh pemanggilan fungsi
    download_geofon_data(
        network="GE",
        station="BBJI",
        location="--",
        channel="BHZ",
        start_time="2024-01-01 00:00:00",
        end_time="2024-01-01 23:59:59",
        output_filename="BBJI_1day.mseed",
        chunk_size=86400,
    )
```

```python
# Baca data
st = read("BBJI_1day.mseed")

# Preprocessing
st.detrend("linear")  # Hilangkan trend linear
st.taper(max_percentage=0.05)  # Aplikasikan taper
st.filter("bandpass", freqmin=0.1, freqmax=20.0)  # Filter bandpass

# Plot waveform
st.plot()
```

```python
# Plot spektogram
st[0].spectrogram(log=True)
```

```python
# Hitung STA/LTA untuk deteksi otomatis
df = st[0].stats.sampling_rate
cft = recursive_sta_lta(st[0].data, int(5 * df), int(10 * df))

# Tentukan threshold
triggers = trigger_onset(cft, 1.5, 0.5)


plt.plot(st[0].data)
for trigger in triggers:
    plt.axvline(trigger[0], color="red", label="Possible P arrival")
plt.show()
```

```python
# Data dan parameter
trace_data = st[0].data
sampling_rate = df  # Sampling rate in Hz

# Hitung PSD
spec, freq = psd(trace_data, NFFT=2048, Fs=sampling_rate)

# Plot
plt.figure()
plt.plot(freq, spec)
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Frequency [Hz]")
plt.ylabel("Power Spectral Density")
plt.title("Power Spectral Density")
plt.show()
```

```python
def fetch_waveform_with_interval(
    network,
    station,
    location,
    channel,
    timestamp,
    interval,
    save_to_file=False
):
    """
    Mengunduh data seismik dari jaringan GEOFON berdasarkan waktu tertentu dan interval.
    Parameters:
        network (str): Kode jaringan (contoh: "GE").
        station (str): Kode stasiun (contoh: "BBJI").
        location (str): Lokasi (contoh: "--").
        channel (str): Kanal (contoh: "BHZ").
        timestamp (str): Waktu dalam format "YYYY-MM-DD HH:MM:SS".
        interval (int): Interval waktu dalam detik dari timestamp.
        save_to_file (bool): Jika True, data akan disimpan ke file lokal.
    Returns:
        Stream: Objek ObsPy Stream yang berisi data yang diunduh.
    """
    # Inisialisasi client GEOFON
    client = Client("GFZ")
    
    # Konversi waktu ke UTCDateTime
    center_time = UTCDateTime(timestamp)
    start_time = center_time - interval
    end_time = center_time + interval
    
    try:
        # Buat progress bar untuk proses download
        with tqdm(total=100, 
                 desc=f"Mengunduh data {network}.{station}", 
                 bar_format='{l_bar}{bar}| {n_fmt}/{total_fmt} [{elapsed}<{remaining}]') as pbar:
            
            # Update progress ke 30% untuk persiapan download
            pbar.update(30)
            
            st = client.get_waveforms(
                network=network,
                station=station,
                location=location,
                channel=channel,
                starttime=start_time,
                endtime=end_time,
            )
            
            if save_to_file:
                # Update progress ke 80% setelah download selesai
                pbar.update(50)
                
                timestamp_str = center_time.strftime("%Y%m%dT%H%M%S")
                output_filename = f"{network}.{station}.{timestamp_str}.mseed"
                st.write(output_filename, format="MSEED")
                
                # Update progress ke 100% setelah file disimpan
                pbar.update(20)
            else:
                # Jika tidak menyimpan file, langsung update ke 100%
                pbar.update(70)
            
            time.sleep(0.05)  # Memberikan waktu untuk melihat progress selesai
            return st
    
    except Exception as e:
        print(f"\nGagal mengunduh data: {str(e)}")
        return Stream()

if __name__ == "__main__":
    # Contoh penggunaan dengan multiple stations
    stations = ["BBJI", "CISI", "GSI"]
    
    for station in stations:
        print(f"\nMencoba mengunduh data untuk stasiun {station}:")
        stream = fetch_waveform_with_interval(
            network="GE",
            station=station,
            location="--",
            channel="BHZ",
            timestamp="2024-01-01 12:00:00",
            interval=3600,  # 1 jam sebelum dan sesudah
            save_to_file=True  # Simpan hasil ke file
        )
        print(f"Stream yang diunduh untuk {station}:")
        print(stream)
        print("-" * 50)
```

```python
def get_earthquake_data(
    start_time,
    end_time,
    minlatitude,
    maxlatitude,
    minlongitude,
    maxlongitude,
    minmagnitude,
    allowed_stations,
):
    """
    Retrieves earthquake event data and associated waveform data for a specified time range and geographic area
    from the GEOFON network. The function fetches both event metadata and waveform data for multiple components
    (BHZ, BHE, BHN) from allowed stations.

    Parameters
    ----------
    start_time : UTCDateTime
        The start time for the earthquake event search.
        Format should be a UTCDateTime object (e.g., UTCDateTime("2024-01-01")).
    
    end_time : UTCDateTime
        The end time for the earthquake event search.
        Format should be a UTCDateTime object (e.g., UTCDateTime("2024-01-02")).
    
    minlatitude : float
        The minimum latitude of the search area in degrees (-90 to 90).
        Negative values represent southern latitudes.
    
    maxlatitude : float
        The maximum latitude of the search area in degrees (-90 to 90).
        Must be greater than minlatitude.
    
    minlongitude : float
        The minimum longitude of the search area in degrees (-180 to 180).
        Negative values represent western longitudes.
    
    maxlongitude : float
        The maximum longitude of the search area in degrees (-180 to 180).
        Must be greater than minlongitude.
    
    minmagnitude : float
        The minimum magnitude threshold for earthquakes to include.
        Events with magnitude less than this value will be excluded.
    
    allowed_stations : set
        A set of station codes to include in the results.
        Only waveform data from these stations will be retrieved.
        Example: {"BBJI", "CISI", "GSI"}

    Returns
    -------
    list
        A list of dictionaries, where each dictionary contains information about
        an earthquake event. Each dictionary has the following structure:
        {
            "time": UTCDateTime,          # Time of the earthquake
            "latitude": float,            # Latitude of epicenter
            "longitude": float,           # Longitude of epicenter
            "depth_km": float,            # Depth in kilometers
            "magnitude": float,           # Earthquake magnitude
            "magnitude_type": str,        # Type of magnitude (e.g., "mb", "ML")
            "stations": [                 # List of stations that recorded the event
                {
                    "station": str,       # Station identifier (e.g., "GE.BBJI")
                    "phase": str          # Phase type (e.g., "P", "S")
                },
                ...
            ],
            "waveforms": {               # Dictionary of waveform data
                "GE.STATION": {          # Station identifier
                    "BHZ": Stream,       # Vertical component
                    "BHE": Stream,       # East-West component
                    "BHN": Stream        # North-South component
                },
                ...
            }
        }

    Examples
    --------
    >>> from obspy import UTCDateTime
    >>> start = UTCDateTime("2024-01-01")
    >>> end = UTCDateTime("2024-01-02")
    >>> stations = {"BBJI", "CISI"}
    >>> results = get_earthquake_data(
    ...     start_time=start,
    ...     end_time=end,
    ...     minlatitude=-11,
    ...     maxlatitude=6,
    ...     minlongitude=95,
    ...     maxlongitude=141,
    ...     minmagnitude=4.0,
    ...     allowed_stations=stations
    ... )
    >>> for event in results:
    ...     print(f"Earthquake at {event['time']}, Magnitude {event['magnitude']}")

    Notes
    -----
    1. The function uses the GEOFON FDSN web service client to fetch data
    2. Waveform data is fetched for three components: BHZ, BHE, and BHN
    3. Time window for waveform data is 600 seconds around the event time
    4. Only stations from the GE (GEOFON) network are considered
    5. If waveform download fails for any station/component, it will be skipped
    6. Phase information includes both P and S waves when available

    Raises
    ------
    Exception
        If there are issues connecting to the GEOFON service or retrieving data.
        Individual waveform download failures are logged but don't stop execution.
    """
    client = Client("GFZ")
    catalog = client.get_events(
        starttime=start_time,
        endtime=end_time,
        minlatitude=minlatitude,
        maxlatitude=maxlatitude,
        minlongitude=minlongitude,
        maxlongitude=maxlongitude,
        minmagnitude=minmagnitude,
        includearrivals=True,
    )

    results = []
    channels = ["BHZ", "BHE", "BHN"]  # Mengambil semua komponen

    for event in catalog:
        origin = event.preferred_origin() or event.origins[0]
        magnitude = event.preferred_magnitude() or event.magnitudes[0]

        event_data = {
            "time": origin.time,
            "latitude": origin.latitude,
            "longitude": origin.longitude,
            "depth_km": origin.depth / 1000,
            "magnitude": magnitude.mag,
            "magnitude_type": magnitude.magnitude_type,
            "stations": [],
            "waveforms": {},
        }

        stations_seen = set()

        for pick in event.picks:
            station = pick.waveform_id.station_code
            network = pick.waveform_id.network_code
            phase = pick.phase_hint
            
            if network == "GE" and station in allowed_stations:
                station_info = f"{network}.{station}"
                station_phase_key = f"{station_info}.{phase}"
                
                if station_phase_key not in stations_seen:
                    stations_seen.add(station_phase_key)
                    event_data["stations"].append({
                        "station": station_info,
                        "phase": phase
                    })

                    # Mengunduh data untuk semua komponen
                    if station_info not in event_data["waveforms"]:
                        event_data["waveforms"][station_info] = {}
                        
                    for channel in channels:
                        try:
                            st = fetch_waveform_with_interval(
                                network=network,
                                station=station,
                                location="--",
                                channel=channel,
                                timestamp=str(origin.time),
                                interval=600,
                                save_to_file=False
                            )
                            if st:
                                event_data["waveforms"][station_info][channel] = st
                        except Exception as e:
                            print(f"Gagal mengambil waveform untuk {station_info}.{channel}: {e}")

        results.append(event_data)

    return results

if __name__ == "__main__":
    start_time = UTCDateTime("2024-01-01")
    end_time = UTCDateTime("2024-01-02")
    
    allowed_stations = {
        "BBJI", "CISI", "GSI", "BKB", "BKNI", "BNDI", "FAKI",
        "GENI", "JAGI", "LHMI", "LUWI", "MMRI", "MNAI", "PLAI",
        "PMBI", "PMBT", "SANI", "SAUI", "SMRI", "SOEI", "TNTI",
        "TOLI", "TOLI2", "UGM", "YOGI"
    }

    results = get_earthquake_data(
        start_time=start_time,
        end_time=end_time,
        minlatitude=-11,
        maxlatitude=6,
        minlongitude=95,
        maxlongitude=141,
        minmagnitude=4.0,
        allowed_stations=allowed_stations,
    )

    # Print results
    for event in results:
        print(f"Waktu: {event['time']}")
        print(f"Lokasi: {event['latitude']:.3f}, {event['longitude']:.3f}")
        print(f"Kedalaman: {event['depth_km']:.1f} km")
        print(f"Magnitude: {event['magnitude']:.1f} {event['magnitude_type']}")
        print("Stasiun yang mencatat:")
        for station in event["stations"]:
            print(f"    - {station['station']} (Phase: {station['phase']})")
        print(f"Total stasiun: {len(event['stations'])}")
        print("Waveforms tersedia untuk stasiun berikut:")
        for station, waveform in event["waveforms"].items():
            print(f"    - {station}: {waveform}")
        print("-" * 50)
```

```python
# Fungsi untuk mendapatkan data stasiun GEOFON
def get_geofon_stations():
    """
    Mengambil data stasiun GEOFON di Indonesia dari layanan FDSNWS.

    Fungsi ini mengunduh data stasiun dari layanan FDSNWS berbasis HTTP, memfilter stasiun
    yang berada di Indonesia berdasarkan daftar `target_stations`, lalu mengembalikan
    data stasiun dalam format DataFrame.

    Returns:
        pandas.DataFrame: Data stasiun GEOFON di Indonesia, terdiri dari kolom:
            - Network: Kode jaringan
            - Station: Nama stasiun
            - Latitude: Garis lintang stasiun
            - Longitude: Garis bujur stasiun
            - Elevation: Ketinggian stasiun
            - SiteName: Nama lokasi stasiun
            - StartTime: Waktu mulai operasional stasiun
            - EndTime: Waktu akhir operasional stasiun (jika ada)

    Raises:
        requests.exceptions.RequestException: Jika terjadi kesalahan saat mengambil data dari server.
    """
    target_stations = [
        "BBJI",
        "CISI",
        "GSI",
        "BKB",
        "BKNI",
        "BNDI",
        "FAKI",
        "GENI",
        "JAGI",
        "LHMI",
        "LUWI",
        "MMRI",
        "MNAI",
        "PLAI",
        "PMBI",
        "PMBT",
        "SANI",
        "SAUI",
        "SMRI",
        "SOEI",
        "TNTI",
        "TOLI",
        "TOLI2",
        "UGM",
        "YOGI",
    ]

    base_url = "http://geofon.gfz-potsdam.de/fdsnws/station/1/query"
    params = {"network": "GE", "format": "text", "level": "station"}

    try:
        response = requests.get(base_url, params=params)
        response.raise_for_status()

        df = pd.read_csv(
            response.url,
            sep="|",
            names=[
                "Network",
                "Station",
                "Latitude",
                "Longitude",
                "Elevation",
                "SiteName",
                "StartTime",
                "EndTime",
            ],
        )

        df_indonesia = df[df["Station"].isin(target_stations)]
        df_indonesia = (
            df_indonesia.sort_values("StartTime")
            .groupby("Station")
            .last()
            .reset_index()
        )

        return df_indonesia

    except requests.exceptions.RequestException as e:
        print(f"Error mengambil data: {e}")
        return None


# Fungsi untuk membuat koneksi fully connected
def create_fully_connected_edges(n_stations):
    """
    Membuat koneksi fully connected untuk sejumlah stasiun.

    Fungsi ini menghasilkan indeks pasangan stasiun (edges) yang saling terhubung
    secara penuh, kecuali koneksi ke diri sendiri.

    Args:
        n_stations (int): Jumlah stasiun.

    Returns:
        numpy.ndarray: Array dua dimensi (2 x jumlah koneksi) yang merepresentasikan
        pasangan indeks (edges) antara stasiun-stasiun.
        - Baris pertama: Indeks awal koneksi
        - Baris kedua: Indeks akhir koneksi
    """
    edges = []
    for i in range(n_stations):
        for j in range(n_stations):
            if i != j:
                edges.append([i, j])
    return np.array(edges).T


# Fungsi untuk memplot jaringan stasiun GEOFON di Indonesia
def plot_network_indonesia(stations_df, edge_index):
    """
    Memvisualisasikan jaringan stasiun GEOFON di Indonesia dengan koneksi fully connected.

    Fungsi ini memplot jaringan stasiun di peta Indonesia menggunakan data
    lokasi stasiun dan koneksi antar-stasiun. Jaringan fully connected ditampilkan
    dengan garis biru yang menghubungkan setiap pasangan stasiun.

    Args:
        stations_df (pandas.DataFrame): Data stasiun GEOFON, dengan kolom:
            - Longitude: Garis bujur stasiun
            - Latitude: Garis lintang stasiun
            - Station: Nama stasiun
        edge_index (numpy.ndarray): Array dua dimensi (2 x jumlah koneksi) yang
        merepresentasikan pasangan indeks (edges) antara stasiun-stasiun.

    Saves:
        File PNG: Menyimpan gambar jaringan dengan nama `geofon_network_indonesia_final_with_map.png`.

    Displays:
        Matplotlib Figure: Visualisasi jaringan stasiun di peta Indonesia.
    """
    stations_df["Longitude"] = pd.to_numeric(stations_df["Longitude"], errors="coerce")
    stations_df["Latitude"] = pd.to_numeric(stations_df["Latitude"], errors="coerce")
    stations_df = stations_df.dropna(subset=["Longitude", "Latitude"])

    plt.figure(figsize=(15, 10))
    ax = plt.axes(projection=ccrs.PlateCarree())
    ax.set_extent(
        [
            stations_df["Longitude"].min() - 5,
            stations_df["Longitude"].max() + 5,
            stations_df["Latitude"].min() - 5,
            stations_df["Latitude"].max() + 5,
        ]
    )

    ax.add_feature(cfeature.LAND, edgecolor="black")
    ax.add_feature(cfeature.COASTLINE)
    ax.add_feature(cfeature.BORDERS, linestyle=":")
    ax.add_feature(cfeature.LAKES, alpha=0.5)
    ax.add_feature(cfeature.RIVERS)

    for i in range(edge_index.shape[1]):
        start = edge_index[0, i]
        end = edge_index[1, i]
        start_lon = stations_df.iloc[start]["Longitude"]
        start_lat = stations_df.iloc[start]["Latitude"]
        end_lon = stations_df.iloc[end]["Longitude"]
        end_lat = stations_df.iloc[end]["Latitude"]

        ax.plot(
            [start_lon, end_lon],
            [start_lat, end_lat],
            color="blue",
            alpha=0.8,
            linewidth=0.1,
            transform=ccrs.PlateCarree(),
        )

    ax.scatter(
        stations_df["Longitude"],
        stations_df["Latitude"],
        c="red",
        s=100,
        label="GEOFON Stations",
        transform=ccrs.PlateCarree(),
    )

    for idx, row in stations_df.iterrows():
        lon = float(row["Longitude"])
        lat = float(row["Latitude"])
        station = str(row["Station"])

        plt.text(
            lon + 0.3,
            lat,
            station,
            fontsize=8,
            bbox=dict(facecolor="white", alpha=0.7, edgecolor="none"),
            transform=ccrs.PlateCarree(),
        )

    ax.set_title("GEOFON Network Indonesia - Fully Connected Network", pad=20)
    ax.legend(loc="upper right")

    # Tambahkan label koordinat pada sumbu
    ax.set_xticks(
        np.arange(
            int(stations_df["Longitude"].min()) - 5,
            int(stations_df["Longitude"].max()) + 6,
            5,
        ),
        crs=ccrs.PlateCarree(),
    )
    ax.set_yticks(
        np.arange(
            int(stations_df["Latitude"].min()) - 5,
            int(stations_df["Latitude"].max()) + 6,
            5,
        ),
        crs=ccrs.PlateCarree(),
    )
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"{x}E"))
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f"{y}S"))

    plt.savefig("geofon_network_indonesia.png", dpi=300, bbox_inches="tight")
    plt.show()


if __name__ == "__main__":
    # Main execution
    stations_df = get_geofon_stations()
    
    if stations_df is not None:
        edge_index = create_fully_connected_edges(len(stations_df))
        np.save("edge_index.npy", edge_index)
    
        print(f"\nJumlah stasiun: {len(stations_df)}")
        print(f"Jumlah koneksi: {edge_index.shape[1]}")
    
        plot_network_indonesia(stations_df, edge_index)
```
